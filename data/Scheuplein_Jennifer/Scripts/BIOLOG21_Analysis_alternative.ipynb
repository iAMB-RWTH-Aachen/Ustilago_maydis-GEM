```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Overview

time = np.array ([48,168,288])

# 48 h = 21.05.2021

A1 = pd.read_excel (r'C:\Users\jenni\Desktop\Forschungspraktikum\Ustilago_maydis-GEM\data\Biolog Daten PM2A Platten.xlsx', sheet_name='21-05-2021')

myA1 = A1.iloc[1:9,1:13].values
myA2 = A1.iloc[11:19,1:13].values
myA3 = A1.iloc[21:29,1:13].values

myA1v = np.ndarray.flatten (myA1)
myA2v = np.ndarray.flatten (myA2)
myA3v = np.ndarray.flatten (myA3)

# 168 h = 26.05.2021
B1 = pd.read_excel (r'C:\Users\jenni\Desktop\Forschungspraktikum\Ustilago_maydis-GEM\data\Biolog Daten PM2A Platten.xlsx', sheet_name='26-05-2021')

myB1 = B1.iloc[1:9,2:14].values
myB2 = B1.iloc[11:19,2:14].values
myB3 = B1.iloc[21:29,2:14].values

myB1v = np.ndarray.flatten (myB1)
myB2v = np.ndarray.flatten (myB2)
myB3v = np.ndarray.flatten (myB3)

# 288 h = 31.05.2021
C1 = pd.read_excel (r'C:\Users\jenni\Desktop\Forschungspraktikum\Ustilago_maydis-GEM\data\Biolog Daten PM2A Platten.xlsx', sheet_name='31-05-2021')

myC1 = C1.iloc[1:9,2:14].values
myC2 = C1.iloc[11:19,2:14].values # empty in this case
myC3 = C1.iloc[21:29,2:14].values

myC1v = np.ndarray.flatten (myC1)
myC2v = np.ndarray.flatten (myC2)
myC3v = np.ndarray.flatten (myC3)

# print (myC1, myC2, myC3)
```

```python
# 48 h mean

myA_mean = np.mean ([myA1v, myA2v, myA3v], axis = 0)
myA_std = np.std ([myA1v, myA2v, myA3v], axis = 0)
myAw_mean = np.reshape (myA_mean, (8,12))
myAw_std = np.reshape (myA_std, (8,12))

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myAw_mean, extent= [1, 12, 8, 1])
plt.title('248h mean')
plt.colorbar()
plt.show()

# 48 h std
myA_mean = np.mean ([myA1v, myA2v, myA3v], axis = 0)
myA_std = np.std ([myA1v, myA2v, myA3v], axis = 0)
myAw_mean = np.reshape (myA_mean, (8,12))
myAw_std = np.reshape (myA_std, (8,12))

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myAw_std, extent= [1, 12, 8, 1])
plt.title('48h std')
plt.colorbar()
plt.show()

# 48 h std/mean

myApercent = myAw_std/myAw_mean

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myApercent, extent= [1, 12, 8, 1])
plt.title('48h std/mean')
plt.colorbar()
plt.show()
plt.hist(myApercent)
plt.title('48h std/mean')
plt.show()

# 168 h mean

myB_mean = np.mean ([myB1v, myB2v, myB3v], axis = 0)
myB_std = np.std ([myB1v, myB2v, myB3v], axis = 0)
myBw_mean = np.reshape (myB_mean, (8,12))
myBw_std = np.reshape (myB_std, (8,12))

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myBw_mean, extent= [1, 12, 8, 1])
plt.title('168h mean')
plt.colorbar()
plt.show()

# 168 h std

myB_mean = np.mean ([myB1v, myB2v, myB3v], axis = 0)
myB_std = np.std ([myB1v, myB2v, myB3v], axis = 0)
myBw_mean = np.reshape (myB_mean, (8,12))
myBw_std = np.reshape (myB_std, (8,12))

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myBw_std, extent= [1, 12, 8, 1])
plt.title('168h std')
plt.colorbar()
plt.show()

# 168 h std/mean

myBpercent = myBw_std/myBw_mean

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myBpercent, extent= [1, 12, 8, 1])
plt.title('288h std/mean')
plt.colorbar()
plt.show()
plt.hist(myBpercent)
plt.title('168h std/mean')
plt.show()

# 288 h mean

myC_mean = np.mean ([myC1v, myC3v], axis = 0) # without C2 (empty)
myC_std = np.std ([myC1v, myC3v], axis = 0) # without C2 (empty)
myCw_mean = np.reshape (myC_mean, (8,12))
myCw_std = np.reshape (myC_std, (8,12))

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myCw_mean, extent= [1, 12, 8, 1])
plt.title('288h mean')
plt.colorbar()
plt.show()

# 288 h std

myC_mean = np.mean ([myC1v, myC3v], axis = 0) # without C2 (empty)
myC_std = np.std ([myC1v, myC3v], axis = 0) # without C2 (empty)
myCw_mean = np.reshape (myC_mean, (8,12))
myCw_std = np.reshape (myC_std, (8,12))

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myCw_std, extent= [1, 12, 8, 1])
plt.title('288h std')
plt.colorbar()
plt.show()

# 288 h std/mean

myCpercent = myCw_std/myCw_mean

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y_ticks = ['A','B','C','D','E','F','G','H']
plt.yticks(y, y_ticks)
plt.imshow(myCpercent, extent= [1, 12, 8, 1])
plt.title('288h std/mean')
plt.colorbar()
plt.show()
plt.hist(myCpercent)
plt.title('288h std/mean')
plt.show()
```
