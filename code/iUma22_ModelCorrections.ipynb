```python
# adding reactions for N-acetylglucosamin metabolism
ID = 'NAGK'
if ID not in model.reactions:
    reaction = Reaction(ID)
    reaction.name = 'N-Acetylglucosamin kinase'
    reaction.lower_bound = 0.
    reaction.upper_bound = 1000.
    reaction.annotation = {'biocyc': 'N-ACETYLGLUCOSAMINE-KINASE-RXN', 'brenda': '2.7.1.59'}
    model.add_reactions([reaction])
    model.reactions.get_by_id('NAGK').build_reaction_from_string('acgam_c + atp_c --> acgam6p_c + adp_c + h_c')

    print(reaction.id)
    print(reaction.reaction)
    print(model.reactions.get_by_id('NAGK').check_mass_balance())
    
ID = 'NAGDA'
if ID not in model.reactions:
    reaction = Reaction(ID)
    reaction.name = 'N-Acetylglucosamin6P deacetylase'
    reaction.lower_bound = 0.
    reaction.upper_bound = 1000.
    reaction.annotation = {'biocyc': 'NAG6PDEACET-RXN', 'brenda': '3.5.1.25'}
    model.add_reactions([reaction])
    model.reactions.get_by_id('NAGDA').build_reaction_from_string('acgam6p_c + h2o_c --> gam6p_c + ac_c')

    print(reaction.id)
    print(reaction.reaction)
    print(model.reactions.get_by_id('NAGK').check_mass_balance())
```

```python
# adding reactions for succinate metabolism
model.reactions.get_by_id('TRANS_Malate_Phosphate_Shuttle').lower_bound = -1000
ID = 'RedoxDiss'
if ID not in model.reactions:
    reaction = Reaction(ID)
    reaction.name = 'NADH Redox dissipation'
    reaction.lower_bound = 0.
    reaction.upper_bound = 1000.
    model.add_reactions([reaction])
    reaction.build_reaction_from_string('4 nadh_c + 2 o2_c --> 2 h2o_c + 4 nad_c')

    print(reaction.id)
    print(reaction.reaction)
    print(model.reactions.get_by_id('NAGK').check_mass_balance())
```
