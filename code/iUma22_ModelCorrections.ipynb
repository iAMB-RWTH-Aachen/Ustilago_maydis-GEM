```python
import re
import pandas as pd
from scipy import optimize

Data = np.array([SubDat,GrDat]).T

print(GrDat)
print(SubDat)

# loading growth sum data
RCFile = '../data/GrowthRates/RatesCompare.xlsx'
RCSheet = 'RateCompare'
RateComp = pd.read_excel(RCFile, sheet_name=RCSheet)
Data = RateComp[['Substrate-uptake-mmol/gCDW/h','Growth-Rate-/h']].values

def GAMEQ(GAMVAL, model=model, Data=Data, BM_ID='BIOMASS_REACTION', ATPID='atp_c', EX_Sub_Act=['EX_glc__D_e']):
    GAMVAL = float(GAMVAL)
    BM_EQ_Ref = model.reactions.get_by_id(BM_ID).reaction
    BM_EQ = re.sub(r'\d+\.\d+\ +{}'.format(ATPID), '{} {}'.format(GAMVAL, ATPID), BM_EQ_Ref)
    SubDat = Data[:,0]
    GrDat = Data[:,1]
    SimGr = list()
    SSR = list()
    with model:
        model.reactions.get_by_id(BM_ID).build_reaction_from_string(BM_EQ)
        for myGlc,myGr in zip(SubDat, GrDat):
            SimGr = TestSubstrate(model, EX_Sub_Act=EX_Sub_Act, EX_Sub_Off=['EX_co2_e'], Rate =myGlc)
            SSR.append((myGr-SimGr)**2)
#     RSS = np.sum((myGr-np.array(SimGr))**2)
#     print(RSS)
    return np.sum(SSR)
minimum = optimize.fmin(GAMEQ,32)
print('optimal GAM: {}'.format(minimum[0]))
```
