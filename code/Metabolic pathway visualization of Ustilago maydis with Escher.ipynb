# Metabolic pathway visualization of Ustilago maydis with Escher

## Conversion of the model iCL1079 from sbml to json

```python
# import cobra.test
from os.path import join
from cobra.io import read_sbml_model
# import escher
from escher import Builder
# import cobra
# from time import sleep
# data_dir = cobra.test.data_dir

# pip install pytest-astropy
```

```python
ModelFile = join('..','model','iCL1079.xml')
model=read_sbml_model(ModelFile)
# cobra.io.save_json_model(model, "iCL1079.json")
model.compartments
```

## Global configuration

escher.rc['never_ask_before_quit'] = True

## The Builder

```python
builder=Builder()
```

```python
# load U. maydis ESCHER map with Pentose phosphate pathway, Glycolysis

Escher_File1 = join('..','model','U.maydis_ESCHER_map_PPPW_Glycolysis_TCA.json')
Escher_File2 = join('..','model','U. maydis_ESCHER_map_Glycine_metabolism.json')
builder = Builder(
    map_json=Escher_File1,
    model = model, # 'iCL1079.json',
)
# Run FBA with the model and add the flux data to the map
solution = builder.model.optimize()
builder.reaction_data = solution.fluxes

builder
```
